```python
Практическое задание.
Для того, чтобы закрепить материал, посчитайте значение среднеквадратичной
ошибки для данных наборов точек и аппроксимирующих функций. Постройте графики
этих функций и отметьте данные точки на нем, чтобы посмотреть, как проходит через них
график аппроксимирующей функции. При решении можете пользоваться python и
matplotlib, а можете по-старинке, сделать это на бумаге, хотя это будет не слишком
удобно.
Начальный уровень.
1. (2; 12.258), (4; 17.24), (8; 30.151).
f(x) = 3x + 5.
2. (2; 3.688), (4; 10.791), (8; 20.705).
f(x) = 0.25x .75x .25.
2 + 0 + 1
Средний уровень.
1. (2; 4.872), (4; 29.707), (8; 246.971), (10; 485.727)
```

```python
%matplotlib inline

from sympy import*
from sympy.solvers.solveset import linsolve
from sympy.solvers.inequalities import solve_univariate_inequality
import matplotlib.pyplot as plt
import numpy as np

init_printing(use_unicode = False, wrap_line = False, no_global=True)
```

```python
#Task3
x,y = symbols("x,y")

xv = [2, 4, 8, 10]
yv = [4.872, 29.707, 246.971, 485.727]

f1 = 0.5*x**3 + 0.75*x + 1.25
fx = []
for i in range(len(xv)):
    fx.append(f1.subs(x, xv[i]))

poly_x = []
poly_y = []
for i in range(0,100):
    cur_x = i*11/100 
    poly_x.append(cur_x)
    poly_y.append(f1.subs(x, cur_x))


plt.scatter(xv, yv, label= "values", color= "green", s=60)
plt.plot(poly_x, poly_y, label = "approx") 
plt.legend()
plt.show()

sum_value = 0
for i in range(len(xv)):
    sum_value += (yv[i] - fx[i])**2
mse = sum_value / len(xv)
rmse = mse**0.5
print("mse=", round(mse,2), "; rmse=", round(rmse,2))
```
