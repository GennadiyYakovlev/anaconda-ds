```python
import requests
import pandas as pd

url = "https://itunes.apple.com/search"
params = {'media':'music', 'term':'manizha', 'limit':30}     #"term=manizha&media=music&limit=30"
response = requests.get(url, params)
```

```python
response.status_code
```

```python
result = response.json()['results']
df = pd.DataFrame(result)
df
```

```python
df.describe()
```

```python
df.primaryGenreName.value_counts()
```

```python
df.to_excel('./data/iTunes_api.xlsx', sheet_name='Manizha', index_label='id')
```

```python
df.to_csv('./data/iTunes_api.csv', sep='\t', index=False)
```
